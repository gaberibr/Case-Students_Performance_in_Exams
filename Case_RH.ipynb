```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(), cbar=False);
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,6))
plt.title('Análise de renda')
sns.kdeplot(data['Salary'], color='green');
```

```python
plt.figure(figsize=(10,6))
plt.title('Análise Experiência')
sns.kdeplot(data['YearsExperience'], color='blue');
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data['Salary']);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x='Salary', y='YearsExperience');
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=data, x='Salary', y='YearsExperience');
```

```python
correlacao = data.corr('pearson')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao, annot=True, cbar=True);
```

```python
correlacao = data.corr('spearman')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao, annot=True, cbar=True);
```
