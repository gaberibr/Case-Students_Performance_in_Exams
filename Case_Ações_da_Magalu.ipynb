```python
#bibliotecas para modelagens e matrizes
import numpy as np
import pandas as pd

#bibliotecas para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

#biblioteca para ignorar avisos
import warnings
#desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#leitura de dados
data_base = pd.read_excel('Magalu.xlsx')

#verificando dados
data_base.head()
```

```python
data_base.shape
```

```python
data_base.info()
```

```python
data_base.describe()
```

```python
#séries temporais
Dados = data_base.set_index('Data')
Dados.head()
```

```python
#registro de evento durante um período
plt.style.use('seaborn-dark')
plt.figure(figsize=(16,5))
plt.title('Análise de ações - Magalu', fontsize=14, loc='left')
plt.plot(Dados.index, Dados['Fechamento'])

plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da ação (R$)');
```

entender o que aconteceu.
as ações estavam sendo negociadas em torno de 25 reais. Depois de março teve uma queda, subiu, depois de maio, despencou, e subiu novamente. E posteriormente teve um grande declínio.

Análisando esse gráfico. Investiria na Magalu?
  como as ações estão baixa, pode ser uma boa oportunidade para comprar, pois algum periodos já foram altos.

```python
Dados.tail()
```

```python
#média móvel
media_movel = Dados['Fechamento'].rolling(5).mean()
media_tendencia = Dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-dark')
plt.figure(figsize=(16,5))
plt.title('Análise de ações - Magalu', fontsize=14, loc='left')

plt.plot(Dados.index, Dados['Fechamento'])
plt.plot(media_movel.index, media_movel)
plt.plot(media_tendencia.index, media_tendencia)

plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da ação (R$)');
```

```python
sns.boxplot(data=Dados, x='Fechamento');
```

```python
#boxplot mensal
data_base['Mês'] = data_base['Data'].dt.month

plt.figure(figsize=(16,5))
sns.boxplot(data=data_base, x='Mês', y='Fechamento');
```

```python
data_base.groupby(['Mês']).describe()['Fechamento']
```

```python
#plot
grafico = go.Figure(
    data=[
        go.Candlestick(
          x= Dados.index,
          open = Dados['Abertura'],
          high = Dados['Maior'],
          low = Dados['Menor'],
          close = Dados['Fechamento'],
        )
    ]
)

grafico.update_layout(xaxis_rangeslider_visible=False)
grafico.show()
```
