```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('Dados_Pib.xlsx')
data.head(10)
```

```python
data.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
#Sistema de grids
background = '#f5f5f5'

grid_graficos = sns.FacetGrid(data, col='Territorialidades', hue='Territorialidades', col_wrap=4)

grid_graficos = grid_graficos.map(plt.plot, 'Ano', 'PIB per capita')

grid_graficos = grid_graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

grid_graficos = grid_graficos.set_titles('{col_name}')

grid_graficos = grid_graficos.fig.suptitle(
    'Evolução da renda per capita por estado', fontsize=18
)

plt.subplots_adjust(top=0.92);
```
