```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Empresas_Energia.xlsx')

data.head()
```

```python
data.set_index('Data', inplace=True)

data.head()
```

```python
# Grafico

# Estilo do gráfico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure( figsize=(15, 5))

# Titulo
plt.title('Análise de ações - Empresas de Energia', loc='left', fontsize=18, fontweight=0 )

# Plot da Petrobras
plt.plot( data.index, data['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7 )
# Texto da Petrobras
plt.text( data.index[-1], data['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left' )

# Plot de todas as colunas
for coluna in data.columns[1:]:
  # Plot das outras ações
  plt.plot( data.index, data[coluna], color='gray', linewidth=2, alpha=0.7 )
  # Texto das outras ações
  plt.text( data.index[-1], data[coluna].tail(1), coluna, color='gray', size='large', horizontalalignment='left' )

# Labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
